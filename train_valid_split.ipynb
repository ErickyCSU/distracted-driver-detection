```python
import os, shutil
import pandas as pd
import numpy as np
```

```python
# Global variables
base_dir = 'C:\\Users\ERIC\Desktop\graduate project\imgs\\'
```

```python
img_info_df = pd.read_csv('driver_imgs_list.csv')
print(img_info_df.head(5))
```

```python
driver_index = list(set(img_info_df['subject']))
print(driver_index)
```

```python
def train_valid_folder(folder_name):
    folder_path = base_dir+ folder_name
    if os.path.exists(folder_path):
        shutil.rmtree(folder_path)
    os.mkdir(folder_path)
    for i in range(10):
        os.mkdir(folder_path+'\\c'+str(i))
        

def train_valid_split(valid_index):
    train_valid_folder('new_train')
    train_valid_folder('new_valid')
    
    valid_img_name = []
    for index in valid_index:
        if index not in driver_index:
            print("Illegal index!")
            return
        img_list = img_info_df[img_info_df['subject']==index]['img'].values
        valid_img_name.extend(img_list)
    
    img_num = 0
    for i in range(10):
        folder_path = base_dir+'\\train\\c'+ str(i)

        for file_name in os.listdir(folder_path):
            if file_name in valid_img_name:
                new_folder_path = base_dir+'\\new_valid\\c'+ str(i)
            else:
                new_folder_path = base_dir+'\\new_train\\c'+ str(i)
            src = os.path.join(folder_path,file_name)
            des = os.path.join(new_folder_path,file_name)
            shutil.copyfile(src, des)
            img_num += 1
            print('The {}th image is copied'.format(img_num))
            
```

```python
train_valid_split(['p012', 'p041', 'p002'])
```
