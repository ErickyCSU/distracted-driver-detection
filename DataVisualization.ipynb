```python
import matplotlib.pyplot as plt
from PIL import Image
import os, random

random.seed(2018)
base_dir = '.\\imgs\\train'
row_num = 2
col_num = 5
plt.figure(figsize=(8*row_num, col_num))
for i in range(10):
    plt.subplot(row_num, col_num, i+1)
    plt.title('Image from c'+str(i))
    class_dir = base_dir + '\c' + str(i)
    img_dir = class_dir + '\\' + random.choice(os.listdir(class_dir))
    img = Image.open(img_dir)
    plt.imshow(img)
    plt.axis('off')

if not os.path.exists('visualization'):
    os.mkdir('.\\visualization')
plt.savefig('.\\visualization\\vis1')    

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as color
import pandas as pd
import numpy as np

class_num = 10
img_info_df = pd.read_csv('driver_imgs_list.csv')
driver_id_list = list(set(img_info_df['subject'].values))
print(driver_id_list)
```

```python
class_list = ['c{}'.format(i) for i in range(class_num)]
col_num = 4
row_num = 7
plt.figure(figsize=(8*row_num, 15*col_num))
for i, driver_id in enumerate(driver_id_list):
    plt.subplot(row_num, col_num, i+1)
    plt.title('Image number of driver '+driver_id, fontsize=30)
    driver_info_df = img_info_df[img_info_df['subject']==driver_id]
    num_list = [len(driver_info_df[driver_info_df['classname']=='c'+str(i)]) for i in range(class_num)]
    plt.bar(range(len(num_list)), num_list, label=driver_id, tick_label = class_list)
    plt.xticks(fontsize=30)
    plt.yticks(fontsize=30)
    
plt.savefig('.\\visualization\\vis2')    
plt.legend()
plt.show()  
```

```python
class_list = ['c{}'.format(i) for i in range(class_num)]
col_num = 5
row_num = 6
num_list = np.zeros(10)
plt.figure(figsize=(8*row_num, 5*col_num))
for i, driver_id in enumerate(driver_id_list):
    #plt.subplot(row_num, col_num, i+1)
    plt.title('Image number of driver '+driver_id)
    driver_info_df = img_info_df[img_info_df['subject']==driver_id]
    new_num_list = [len(driver_info_df[driver_info_df['classname']=='c'+str(i)]) for i in range(class_num)]
    plt.bar(range(len(num_list)), new_num_list, bottom=num_list, label=driver_id, tick_label = class_list)
    num_list += new_num_list
plt.savefig('.\\visualization\\vis2')    
plt.legend()
plt.show()
```
